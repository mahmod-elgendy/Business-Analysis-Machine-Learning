```python
import pandas as pd
import numpy as np

def collect_business_data():
    print("Welcome to the Business Data Input System!")

    business_type = input("Enter your business type (e.g., retail, restaurant, e-commerce): ")
    time_series = input("Enter the timeline for your data (day1, month1, year1): ").lower()
    all_data = []
    c = 1

    while True:
        c += 1
        print("\nPlease enter the following operational cost data for", time_series + ":")
        salaries = float(input("Salaries: "))
        taxes = float(input("Taxes: "))
        utilities = float(input("Utilities: "))
        maintenance = float(input("Maintenance: "))
        transportation_cost = float(input("Transportation Cost: "))
        transportation_damage = float(input("Transportation Damages: "))
        staff_bonus = float(input("Staff Bonus: "))
        vacations = float(input("Vacations: "))
        insurance = float(input("Insurance: "))
        material_damage = float(input("Material Damage: "))
        device_maintenance = float(input("Device Maintenance: "))
        events = input(f"Any special events or holidays that affect numbers at {time_series}? (yes/no): ").lower()
        if events == "yes":
          events = input("Enter the name of the event or holiday: ")
        if events == "no":
          events = 0

        all_data.append({
            "time": time_series,
            "salaries": salaries,
            "taxes": taxes,
            "utilities": utilities,
            "maintenance": maintenance,
            "transportation_cost": transportation_cost,
            "transportation_damage": transportation_damage,
            "staff_bonus": staff_bonus,
            "vacations": vacations,
            "insurance": insurance,
            "material_damage": material_damage,
            "device_maintenance": device_maintenance,
            "events": events
        })

        more_data = input("Do you want to add data for another time series? (yes/no): ").lower()
        if more_data != "yes":
            break
        time_series = input(f"Enter the next timeline for your data (e.g., day{c}, month{c}, year{c}): ").lower()

    business_data = pd.DataFrame(all_data)
    print("\nBusiness data successfully collected!\n")
    return business_data, business_type


# Function to collect item and event data
def collect_item_and_event_data():
    items = []

    while True:
        # Collect item details
        print("\nEnter item details:")
        item_name = input("Item name: ")
        item_cost = float(input(f"Cost to produce {item_name}: "))
        selling_price = float(input(f"Selling price of {item_name}: "))
        restock_amount = int(input(f"How many of {item_name} do you restock per the selected time series? "))
        units_sold = int(input(f"How many of {item_name} are sold per the selected time series? "))

        # Store item data
        items.append({
            "Item Name": item_name,
            "Cost": item_cost,
            "Selling Price": selling_price,
            "Restock Amount": restock_amount,
            "Units Sold": units_sold,
        })

        # Check if more items need to be added
        more_items = input("Do you want to add another item? (yes/no): ").strip().lower()
        if more_items != "yes":
            break

    # Collect event data
    events = []
    event_input = input("\nDo you have special events or holidays that affect sales? (yes/no): ").strip().lower()
    if event_input == "yes":
        while True:
            event_name = input("\nEnter event name (e.g., weekend, holiday): ")
            event_item_sales = {}
            for item in items:
                item_name = item["Item Name"]
                sold_during_event = int(input(f"How many of {item_name} are sold during {event_name}? "))
                event_item_sales[item_name] = sold_during_event

            # Store event data
            events.append({
                "Event Name": event_name,
                **event_item_sales
            })

            more_events = input("Do you want to add another event? (yes/no): ").strip().lower()
            if more_events != "yes":
                break

    # Create DataFrames
    items_df = pd.DataFrame(items)
    events_df = pd.DataFrame(events) if events else pd.DataFrame()

    print("\nItem and event data successfully collected!")
    return items_df, events_df
```

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.neighbors import KNeighborsClassifier, KNeighborsRegressor
from sklearn.neural_network import MLPClassifier, MLPRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.svm import SVR
from sklearn.ensemble import VotingClassifier, VotingRegressor, RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.preprocessing import StandardScaler

#----------------------------> Needs Change Later
df = pd.read_csv('training.csv')
df2 = pd.read_csv('business.csv')
#----------------------------> Needs Change Later

categories_profit_margins = {
    "retail": 0.35,
    "cafeXrestaurant": 0.65,
    "manufacture": 0.325,
    "Ecommerce": 0.4,
    "services": 0.175,
    "software": 0.85,
    "luxury": 0.7
}


def pre_process(df,df2):
  df = df.fillna(0)
  df2 = df2.fillna(0)
  le = LabelEncoder()
  df['events'] = le.fit_transform(df['events'])
  df2['events'] = le.fit_transform(df2['events'])
  return df, df2


def business_analysis(df, df2): # Needs more data to get accuracte results and formulas enhancement
    df['profit'] = (df['unit_price'] - df['unit_cost']) * df['sold']
    df['revenue'] = df['unit_price'] * df['sold']
    df['revenue'] = df['revenue'].round(2)
    df['profit'] = df['profit'].round(2)

    total_revenue = df['revenue'].sum()
    total_operational_cost = df2[['taxes', 'utilities', 'insurance', 'maintenance', 'transportation_cost']].sum().sum() / 3
    total_employee_cost = df2['insurance'].sum() + df2['taxes'].sum() + df2['utilities'].sum() + df2['maintenance'].sum() + df2['transportation_cost'].sum() + df2['insurance'].sum() + df2['salaries'].sum() + df2['insurance'].sum() + df2['staff_bonus'].sum() + df2['vacations'].sum()

    net_profit = (sum(df['revenue'])/5.83) - total_operational_cost - (total_employee_cost/3)
    profit_margin = (net_profit / total_operational_cost) * 100

    df['total_cost_per_item'] = ((total_operational_cost / df['sold'].sum()) + (total_employee_cost / df['sold'].sum()) + df['unit_cost']).round(2)

    return total_revenue, net_profit, profit_margin, total_operational_cost, total_employee_cost


def evaluate_predictions(y_true, y_pred, label="Model"):
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    mae = mean_absolute_error(y_true, y_pred)
    r2 = r2_score(y_true, y_pred)

    print(f"\n=== {label} Evaluation ===")
    print(f"Mean Squared Error (MSE): {mse:.2f}")
    print(f"Root Mean Squared Error (RMSE): {np.sqrt(mse):.2f}")
    print(f"Mean Absolute Error (MAE): {mae:.2f}")
    print(f"R-Squared Score (R²): {r2:.4f}")

    return {
        "MSE": mse,
        "RMSE": np.sqrt(mse),
        "MAE": mae,
        "R²": r2
    }

def machine_predict(df, df2):
  features_restock = ['unit_cost', 'unit_price', 'sold', 'profit', 'events']
  features_newSold = ['product_pricing', 'sold', 'events']

  X_restock = df[features_restock]
  y_restock = df['restock']
  X_newSold = df[features_newSold]
  y_newSold = df['sold']

  scaler = StandardScaler()
  X_restock_scaled = scaler.fit_transform(X_restock)
  X_newSold_scaled = scaler.fit_transform(X_newSold)

  # Split into training and test sets
  X_train_restock, X_test_restock, y_train_restock, y_test_restock = train_test_split(
      X_restock_scaled, y_restock, test_size=0.1, random_state=42)
  X_train_newSold, X_test_newSold, y_train_newSold, y_test_newSold = train_test_split(
      X_newSold_scaled, y_newSold, test_size=0.1, random_state=42)

  mlp = MLPRegressor(hidden_layer_sizes=(50, 50),
                   activation='relu',
                   solver='adam',
                   max_iter=5000,
                   alpha=0.00001,
                   learning_rate='adaptive')
  linear = LinearRegression()
  svr = SVR()

  voting_regressor = VotingRegressor(estimators=[ ('mlp', mlp), ('linear', linear), ('svr', svr)])
  voting_regressor.fit(X_train_restock, y_train_restock)

  #----------------> Needs change (predict the data for the coming timeseries not the test)
  y_pred_restock = np.ceil(voting_regressor.predict(X_test_restock))
  #----------------> Needs change (predict the data for the coming timeseries not the test)
  print(y_pred_restock)


  voting_regressor = VotingRegressor(estimators=[ ('mlp', mlp), ('linear', linear), ('svr', svr)])
  voting_regressor.fit(X_train_newSold, y_train_newSold)

  #----------------> Needs change (predict the data for the coming timeseries not the test)
  y_pred_newSold = np.floor(voting_regressor.predict(X_test_newSold))
  #----------------> Needs change (predict the data for the coming timeseries not the test)
  print(y_pred_newSold)
  # Evaluate Restock Predictions
  print("\nEvaluating Restock Predictions:")
  evaluate_predictions(y_test_restock, y_pred_restock, label="Restock Prediction")

    # Evaluate New Sold Predictions
  evaluate_predictions(y_test_newSold, y_pred_newSold, label="New Sold Prediction")

  return y_pred_restock, y_test_restock, y_pred_newSold, y_test_newSold
```

```python
def main():
    df = pd.read_csv("training.csv")
    df2 = pd.read_csv("business.csv")
    df, df2 = pre_process(df, df2)

    # Perform business analysis
    print("\n=== Business Analysis ===")
    total_revenue, net_profit, profit_margin, total_operational_cost, total_employee_cost = business_analysis(df, df2)
    print(f"Net Profit: {net_profit}")
    print(f"Profit Margin: {profit_margin:.2%}")
    print(f"Total Operational Cost: {total_operational_cost}")
    print(f"Total Employee Cost: {total_employee_cost}")

    # machine learning predictions
    print("\n=== Running Machine Learning Predictions ===")
    y_pred_restock, y_test_restock, y_pred_newSold, y_test_newSold = machine_predict(df, df2)

    evaluate_predictions(y_test_restock, y_pred_restock, "Restock Prediction")
    evaluate_predictions(y_test_newSold, y_pred_newSold, "New Sold Prediction")

if __name__ == "__main__":
    main()
```
